```python
## Importing the libraries needed
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('C:\Samuelizz\Portfolio Projects\Raw Data\covidsurvey.csv')
```

```python
data.head()
```

```python
## Checking if the data contains any null values
data.isnull().sum()
```

```python
#removing the null values
data.dropna(inplace = True)\

#confirm the changes made
data.isnull().sum()
```

```python
data.info()
```

```python
## Describing the data
data.describe()
```

```python
## What was the students online class experience like?

ratings = data['Rating of Online Class experience'].value_counts().reset_index()
sns.barplot(data = ratings, x = 'Rating of Online Class experience', y = 'index')
plt.xlabel('Number of response')
plt.ylabel('')
plt.title('Ratings of Online class')
plt.show()
```

#### The above table describe the data showing the average values to the measured catgories

```python
## The most medium for the online class
Medium = data['Medium for online class'].value_counts().reset_index()
sns.barplot(data = Medium, x = 'Medium for online class', y = 'index')
plt.xlabel('Number of response')
plt.ylabel('')
plt.title('Most used medium for online class')
plt.tight_layout()
```

#### laptop/Desktop is the most used medium for online class closely followed by Smart Phone

```python
#Checking if there is any correletion between time spent on online class and time spent on self study

sns.lmplot(data = data, x = 'Time spent on Online Class' , y = 'Time spent on self study')
plt.title('corrr between time spent on online class and self study')
plt.show()
plt.tight_layout()
```

#### The time spent on self study is closely independent of the time spent on online class

```python
# correlation between time spent on fitness and time spent on online class
sns.lmplot(data = data, x = 'Time spent on self study', y = 'Time spent on fitness')
plt.title('corrr between time spent on self study and fitness')
plt.show()
```

```python
## correlation between Time spent on Social media and Time spent on self study
sns.lmplot(data = data, x = 'Time spent on social media', y = 'Time spent on self study')
plt.title('corrr between time spent on self study and social media')
plt.show()
```

```python
## correlation between Time spent on sleepbb and Time spent on self study
sns.lmplot(data = data, x = 'Time spent on self study', y = 'Time spent on sleep')
plt.title('corrr between time spent on self study and sleep')
plt.show()
```

```python
## replacing some response in the Time spent on TV column
data['Time spent on TV'].replace({'No tv': '0', 'N': '0', ' ' : '0', 'n' : '0'}, inplace = True)

##Changing the data type of the column
data['Time spent on TV'] = data['Time spent on TV'].astype('float64')

## correlation between Time spent on TV and Time spent on self study
sns.lmplot(data = data, x = 'Time spent on self study', y = 'Time spent on TV')
plt.title('corr between time spent on self study and TV')
plt.show()
```

```python
## correlation between Time spent on TV and Time spent on online class
sns.lmplot(data = data, x = 'Time spent on Online Class', y = 'Time spent on TV')
plt.title('corr between time spent on online class and TV')
plt.show()
```

```python
#The most preferred social media platform
fig = plt.figure(figsize = (6,8))
social_media = data['Prefered social media platform'].value_counts().reset_index()
sns.barplot(data = social_media, y = 'index', x='Prefered social media platform', orient = 'h')
plt.xlabel('Number of response')
plt.ylabel('')
plt.title('Most prefered social media platform')
plt.tight_layout()
```

```python
weight_changes = data.groupby('Change in your weight')['Number of meals per day'].mean().reset_index()
sns.barplot(data = weight_changes, x = 'Change in your weight', y='Number of meals per day')
plt.ylabel('average meal per day')
plt.xticks(rotation = 'horizontal')
plt.show()
plt.tight_layout()
```

##### Of course! Those who take more meal per day gained more weight

```python
data['Health issue during lockdown'].value_counts(normalize = True)
```

##### Good News! about 14% of the participants had health issues during the lockdown while about 86% stayed had none

```python
## The top 10 stress busters during the lockdown
fig = plt.figure(figsize = (6,8))
stress = data['Stress busters'].value_counts().nlargest(10).reset_index()
sns.barplot(data = stress, x = 'Stress busters', y = 'index')
plt.xlabel('number of response')
plt.ylabel('')
plt.title('Top 10 stress busters')
plt.show()
```

##### Music leads the chart, followed by Online gaming, then Watching web series

```python
# what do they miss most?
fig = plt.figure(figsize = (6,5))
miss = data['What you miss the most'].value_counts().nlargest(10).reset_index()
sns.barplot(data = miss, y = 'index', x = 'What you miss the most', orient = 'h')
plt.xlabel('number of response')
plt.ylabel('')
plt.title('What do you miss the most?')
plt.tight_layout()
plt.show()
```

##### Definetely, as a student, you'd miss your school/college more

```python
#connections with family/friends during the lockdown
data['Do you find yourself more connected with your family, close friends , relatives  ?'].value_counts(normalize=True)
```

##### Good to know that majority got connected with their family during the lockdown

```python
# was most of the time utilized during the lockdown?
data['Time utilized'].value_counts(normalize = True)
```

##### The number of participants who utililized their time during the lockdown is as many as those who did not. Interesting!

```python
## summary of the measured data by region of residence
data.groupby('Region of residence')[['Age of Subject', 'Time spent on Online Class', 'Time spent on self study',
                                     'Time spent on fitness', 'Time spent on sleep', 'Time spent on TV',
                                     'Number of meals per day']].mean()
```
